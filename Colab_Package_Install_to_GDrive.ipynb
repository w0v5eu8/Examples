# Colab Package Install to GDrive


## Symbolic link 설정

*   Colab Notebooks 경로의 space 가 있어서 그냥 pip하면 문제가 생김
*   그래서 symbolic link 를 사용

````python
# Goolge Drive 연결
import os, sys  
from google.colab import drive  
drive.mount('/content/drive')
````

symbolic link 설정

```
# symbolic link 설정
# 'Colab Notebooks' 루트 경로를 'SymLink_Colab_Notebooks'로 연결
os.symlink('/content/drive/MyDrive/Colab Notebooks', '/content/drive/MyDrive/SymLink_Colab_Notebooks')
```



## Packages 설치
*   dependency 오류가 나는 것은 일단 정상임.
*   패키지 import, running시 문제가 없는지는 확인해보기


````python
# 프로젝트별로 폴더 구별
dev_name = 'Test' # Dev Project Name: Test
# Colab Notebooks 폴더 아래에 외부 패키지가 설치될 External_Packages 폴더가 미리 존재 해야 함
pg_path = '/content/drive/MyDrive/SymLink_Colab_Notebooks/[Packages]/'+dev_name

# 앞에서 만든 symbolic link를 system path에 추가 
sys.path.append(pg_path) #  systemp path에 추가 방법 1
# sys.path.insert(0, pg_path) #  systemp path에 추가 방법 2 (append는 sys.path 리스트 맨 아래에 추가 insert는 지정 인덱스에 삽입)

# package 설치
!pip install --target=$pg_path MNE
!pip install --target=$pg_path backtrader
!pip install --target=$pg_path backtesting
!pip install --target=$pg_path deap
!pip install --target=$pg_path talib-binary
````

## Package 사용

패키지를 사용하려면 아래 처럼 패키지가 설치된 경로를 symbolic link 경로로 sys.path에 추가하여야 함.
```
# 프로젝트별로 폴더 구별
dev_name = 'Test' # Dev Project Name: Test
# Colab Notebooks 폴더 아래에 외부 패키지가 설치될 External_Packages 폴더가 미리 존재 해야 함
pg_path = '/content/drive/MyDrive/SymLink_Colab_Notebooks/[Packages]/'+dev_name

# 앞에서 만든 symbolic link를 system path에 추가 
sys.path.append(pg_path) #  systemp path에 추가 방법 1
# sys.path.insert(0, pg_path) #  systemp path에 추가 방법 2 (append는 sys.path 리스트 맨 아래에 추가 insert는 지정 인덱스에 삽입
```
