```python
import torch
import numpy as np
import matplotlib.pyplot as plt
from BXAI_Pytorch_ESN import ESN 
import pandas as pd
```

```python
data = np.loadtxt('BXAI_ESN_Minimal_MackeyGlass_t17.txt')
esn=ESN(1,1500,zero_per =0.9996)
esn.fit(data[0:1000])
```

```python
pre = esn.predict(data[1000:2000])
plt.plot(pre.T[:].detach().numpy(),'#ff6347')
plt.plot(data[1000:2000],'b')
plt.show()
```
